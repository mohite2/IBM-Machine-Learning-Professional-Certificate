```python
import sys, random 
print('*** OLD SCHOOL GAME ****', "\n")

print('*** ROCK, PAPER, SCISSOR ****', "\n")

counter = 1 
win = 0 
loss = 0 
tie= 0

while True: 
    print("\n ********* Turn {} ************".format(counter)) ## game loop 

    print("\nTotal wins: {}, losses: {}, ties: {} for the game \n".format(win,loss,tie)) ## game loop 
    if counter == 5: 
        break
    while True: ## Player loop
        print("Give your input (in letter): r = Rock, p = Paper, s = Scissor \n " )
        player = input().lower()
        playermove = player
#         print("\nPlayer Move = {} \n ".format(playermove))
        
        if playermove == 'q':
            sys.exit()
        elif playermove == 'r' or playermove == 'p' or playermove == 's': 
            break
        else: 
            print("Please select valid input: r , p, s, q \n")
###what player chooses: 
    if playermove == 'r': 
        print("\nPlayer choice: Rock")
    elif playermove == 'p': 
        print("\nPlayer choice: Paper")
#         print("Paper versus....")
    elif playermove == 's': 
        print("\nPlayer choice: Scissor")
#         print("Scissor versus....")

###what computer chooses: 
    randomnum = random.randint(1,3)
    
    if randomnum == 1: 
        computermove = 'r'
        print("Computer choice: Rock \n")
    elif randomnum == 2: 
        computermove = 'p'
        print("Computer choice: Paper \n")
    elif randomnum == 3:
        computermove = 's'
        print("Computer choice: Scissor \n")

    if playermove == computermove: 
        print("IT'S A TIE !! PLAY AGAIN")
        tie = tie + 1
    elif playermove == 'r' and computermove == 'p': 
        print("YOU LOSE :( ")
        loss = loss + 1        
    elif playermove == 'r' and computermove == 's': 
        print("YOU WIN")
        win = win + 1
    elif playermove == 'p' and computermove == 'r': 
        print("YOU WIN")    
        win = win + 1
    elif playermove == 'p' and computermove == 's': 
        print("YOU LOSE :( ")    
        loss = loss + 1
    elif playermove == 's' and computermove == 'r': 
        print("YOU LOSE :( ") 
        loss = loss + 1
    elif playermove == 's' and computermove == 'p': 
        print("YOU WIN") 
        win = win + 1
    counter = counter + 1    

if loss > win: 
    print("COMPUTER WINS!!!")
elif win > loss:
    print("CONGRATUALTIONS!! YOU WIN THIS GAME")
```
